####  The mean is the average or the most common value in a collection of numbers.

### Median is a statistical measure that determines the middle value of a dataset listed in ascending order 

#### to understand the median, lets consider a sample value set {1, 2, 3, 4, 5, 50}

#### without the outlier the mean of the dataset is 3 and with outlier 23.3, so to reduce the possibility of error, we use the median.

#### median takes middle number as its value, like if the total count of the dataset is an even number than the median will be average of the 2 middle numbers here in this case 3,4 are middle numbers the average is 3.5.  if the total count of the dataset is odd, than the middle number of dataset is directly considered as the median value

#### by using the median, we can reduce the errors in averages, which happens because of the outliers
