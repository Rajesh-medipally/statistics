### What is an outlier ?
An outlier is a data point in a data set that is distant from all other observations. A data point that lies outside the overall distribution of the dataset.

## What are the criteria to identify an outlier?
#### Data point that falls outside of 1.5 times of an interquartile range above the 3rd quartile and below the 1st quartile
#### Data point that falls outside of 3 standard deviations. we can use a z score and if the z score falls outside of 2 standard deviation

#### generally, the outliers considered to be in the list of maximum or minimum values of a dataset
#### for ex consider dataset {1,5,6,7,8,9,10,11,25} 
#### in this variable the 1 and 25 can be considered as outliers
#### if we consider the dataset as 100% with its values if we want 50% of maximum values we cut from 50% to ascending side from the median like {9,10,11,25}

# Detecting outlier using Z score
# Using Z score
# Formula for Z score = (Observation — Mean)/Standard Deviation

# z = (X — μ) / σ

## here, the standard deviation nothing but the maximum level of variables covered number usually it considered 3 as per the normal distribution

# by using z score

```python
outliers = []
def detect_outliers( dataset, standard_deviation) :
     threshold=standard_deviation
     mean = np.mean(dataset)
     std =np.std(dataset)
    
    
     for i in dataset:
        z_score= (i - mean)/std 
        if np.abs(z_score) > threshold:
            outliers.append(i)
     return outliers
    
```

```python
outliers = detect_outliers(dataset,3)
```

# by using the IQR 

## to use IQR method we have some steps to do

## -> place variable in acsending order
## -> find the first and third quartile  q1,q3
## -> find the inter quartile range ( means to get the middle value of the dataset if we remove 25 from 75 it shows the values in between) IQR= q3-q1
## -> find the lower bound and upper bound, means IQR value will give the variable range, but it can not know the starting point, so by multiplying it with half of total means 1.5 will make the variable value at the near point, ( honeslt stats people came up with this number so just have to remember it)

### any thing that stays below or above this lower or upper bound limits those are outliers

```python
def detect_outliers_by_IQR( dataset) :
    ## sorting the dataset
    s_dataset=sorted(dataset)
    ## finding the q1 and q3 with respect to percentiles 25,75
    quantile1, quantile3 = np.percentile(dataset,[25,75])
    ## finding the IQR range
    iqr_range = quantile3 - quantile1
    lower_bound_val = quantile1 -(1.5 * iqr_range) 
    upper_bound_val = quantile3 +(1.5 * iqr_range)
    return lower_bound_val,upper_bound_val
```

```python
detect_outliers_by_IQR(dataset)
```
